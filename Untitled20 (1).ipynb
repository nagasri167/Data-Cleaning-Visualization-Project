```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create / Load Dataset
data = {
    "Name": ["Ravi", "Anu", "Kiran", "Ravi", "Sita", "Ram", "Kiran", None],
    "Age": [23, 25, np.nan, 23, 29, 30, 27, 22],
    "Salary": [50000, 60000, 55000, 50000, 65000, 700000, 55000, 48000],
    "Department": ["IT", "HR", "IT", "IT", "Finance", "HR", "IT", "Finance"]
}

df = pd.DataFrame(data)

print("Original Data")
print(df)

# Step 2: Check Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# Step 3: Handle Missing Values
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Name"].fillna("Unknown", inplace=True)

# Step 4: Remove Duplicates
df.drop_duplicates(inplace=True)

print("\nCleaned Data")
print(df)

# Step 5: Detect Outliers using Boxplot
plt.figure()
sns.boxplot(x=df["Salary"])
plt.title("Salary Outliers")
plt.show()

# Step 6: Age Distribution
plt.figure()
plt.hist(df["Age"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Step 7: Salary by Department
plt.figure()
sns.barplot(x="Department", y="Salary", data=df)
plt.title("Salary by Department")
plt.show()

# Step 8: Correlation Heatmap
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
